# UMAPS

```python
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="0" # using the first gpu, for most cases, you wouldn't need to change it.

import scanpy as sc
import numpy as np
import pandas as pd
import anndata
import cupy as cp
import cupyx as cpx
import time
import rapids_singlecell as rsc
from rapids_singlecell.cunnData import cunnData
from tqdm.auto import tqdm, trange
import warnings
warnings.filterwarnings("ignore")
import rmm
from rmm.allocators.cupy import rmm_cupy_allocator
rmm.reinitialize(
    managed_memory=True, # Allows oversubscription
    pool_allocator=True, # default is False
    devices=0, # GPU device IDs to register. By default registers only GPU 0.
)

cp.cuda.set_allocator(rmm_cupy_allocator)
```

```python
indir = "/n/holylfs06/LABS/buenrostro_lab/Lab/conrad_rachel_shared_data/final_minimorf/20260217_atacrna_chromvar_v2/h5ads"
```

```python
h5ad_files = [os.path.join(indir, f)
              for f in os.listdir(indir)
              if f.endswith(".h5ad")]

# Parameters
n_neighbors = 50
distance_input = 0.1
louvainres = 1
```

```python
for filein in h5ad_files:
    print(f"Processing: {filein}")
    basename = os.path.splitext(os.path.basename(filein))[0]
    adata = sc.read_h5ad(filein)
    n_neighbors=50
    distance_input=0.1
    louvainres=1
    outdir = "/n/holylfs06/LABS/buenrostro_lab/Lab/conrad_rachel_shared_data/final_minimorf/20260217_atacrna_chromvar_v2/umaps/"
    sc.settings.figdir = outdir
    
    rsc.pp.neighbors(adata, metric='cosine', n_neighbors=n_neighbors)
    rsc.tl.umap(adata, min_dist=distance_input)
    rsc.tl.louvain(adata, resolution=louvainres)
    sc.pl.umap(adata, ncols=1, color=['louvain'], save=f"_{basename}.png")
    
    
    adata.obs['barcode'] = adata.obs_names
    adata.obs['umap_x'] = adata.obsm['X_umap'][:, 0]
    adata.obs['umap_y'] = adata.obsm['X_umap'][:, 1]
    outfile = os.path.join(outdir, f"{basename}.umap.txt")
    
    adata.obs.to_csv(outfile, sep='\t', index=False)
    coo_matrix = adata.obsp['distances'].tocoo()
    row = coo_matrix.row
    col = coo_matrix.col
    mtx = [[] for i in range(len(adata))]
    for r,c in zip(tqdm(row), col):
        mtx[r].append(c)
    
    mtx = np.array(mtx)
    mtx.shape
    mtx = pd.DataFrame(mtx)
    outfile = os.path.join(outdir, f"{basename}.knn.txt")
    mtx.to_csv(outfile, sep='\t', index=False)
```
