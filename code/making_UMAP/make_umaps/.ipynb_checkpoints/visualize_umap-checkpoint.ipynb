```r
umap <- fread("/mnt/users/rachelsavage/paper_minimorf/data/umaps/umapcoor_leiden.100topics.50nn.0.15dist.txt")
```

```r
meta.df <- fread("/mnt/users/rachelsavage/paper_minimorf/data/frequent_files/meta.dedoublet.csv")
```

```r
umap.df <- umap %>% left_join(meta.df) 
set.seed(123)
umap.df <- umap.df %>% slice_sample(n = nrow(umap.df))
```

```r
col_vector <- c(brewer.pal(n = 12, name = "Paired"), 
             brewer.pal(n = 9, name = "Set1"),
             brewer.pal(n = 8, name = "Accent"),
             brewer.pal(n = 8, name = "Dark2")
            )
n <- length(unique(umap.df$louvain))
set.seed(3)
distinct_colors <- sample(col_vector, n)
names(distinct_colors) <- unique(umap.df$louvain)
```

```r
options(repr.plot.width = 14, repr.plot.height = 10)

ggplot(umap.df, aes(x=umap_x, y=umap_y, color = as.factor(louvain)))+
geom_scattermore(alpha=0.15)+
xlab("UMAP X")+
ylab("UMAP Y")+
coord_fixed()+
scale_color_manual(values = distinct_colors, name = "Louvain Cluster")+
theme_void()+
theme(panel.border = element_rect(color = "black", fill = NA, linewidth = 0.5))+
  guides(color = guide_legend(
    override.aes = list(shape = 15, size = 10, alpha = 1)  # 15 = square
  ))
```
